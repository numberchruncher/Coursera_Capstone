<h1>Notebook Segmenting & Clustering Neighborhoods in Toronto<h1>

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

summary_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(summary_url)
summary_soup = BeautifulSoup(page,'html')
summary_table = summary_soup.find('table',{'class':'wikitable sortable'})
array = []
rows = summary_table.findAll('tr')
#header = [col.text for col in rows[0].findAll('th')]

for row in rows[1:-1]:
    tmp_row = []
    for column in row.findAll('td'):
        tmp_row.append(column.text)
    array.append(tmp_row)
df_raw = pd.DataFrame(array, columns=['Postcode', 'Borough', 'Neighbourhood'])
df_raw.head()
```

remove column break from the end of the neighbourhood

```python
df = df_raw.replace('\n','', regex=True)
df.head()
```

excluding not assigned Boroughs

```python
df2 = df.loc[df['Borough'] != 'Not assigned']
df2.head()
```

```python
df3= df2.loc[df2['Neighbourhood'] == 'Not assigned'] 
df3
```

copy Borough to Neighbourhood if Neighbourhood = Not assigned

```python
df3=df2
df3.loc[df3.Neighbourhood == 'Not assigned','Neighbourhood']=df3.Borough
df3
```

<br>creates a series for Neighbourhood based on Postcode and Borough group by ; join all these rows into one and separate them with an ',' </br>
result is a data series (=data frame with one column)

```python
df4=df3.groupby(['Postcode','Borough'])['Neighbourhood'].transform(lambda x:', '.join(x))
df4
```

merge series and dataframe together based on index

```python
df5= df3.merge(df4, left_index= True, right_index = True)
#now there are two Neighborhoods (_x and _y), delete the old one, rename the other;
del df5['Neighbourhood_x']
df5.rename(columns = {'Neighbourhood_y':'Neighbourhood'}, inplace = True)
df5
```

remove duplicates <br/> 
sort by column <br/> 
new index drop = True otherwise you get a new column with the index;

```python
#remove duplicates;
df6=df5.drop_duplicates()
#sort by column
df6=df6.sort_values('Postcode')
#new index drop = True otherwise you get a new column with the index;
df6=df6.reset_index(drop = True)
df6
```

<br> Shape of dataframe as requested</br>

```python
df6.shape
```

create a list of all postcodes

```python
liste_postcode = []
for pc in df6['Postcode']:   
    liste_postcode.append(pc) 
liste_postcode
```

pypostalcode is a postalcodedatabase  <br/> 
empty list is generated  <br/> 
then loop through list and get a list of postcode, latitude & longitude

```python
from pypostalcode import PostalCodeDatabase
pcdb = PostalCodeDatabase()
list_complete = []
for item in liste_postcode:
    try:
        location = pcdb[item]
        list_complete.append([item,location.latitude, location.longitude])
    except:
         list_complete.append([item,1, 1])
    
```

```python
list_complete
```

change list to dataframe <br/> 
get columns<br/> 
create list with missing values<br/>
delete entry out of list with missing value

```python
#change list to dataframe
df_locations = pd.DataFrame(list_complete)
#add header
df_locations.columns = ['Postcode_list', 'Latitude', 'Longitude']
df_locations.head()
#create list with missing value
MR7 = ['M7R',45.4204,-75.6924]
#'M7R' is not in the database therefore insert this manually
#first delete entry;
#delete entry that has no coordinates
# ~ means not in
df_locations2=df_locations[~df_locations['Postcode_list'].isin(MR7)]
df_locations2
```

<br>sort list and check if entry is really gone</br>
<br>now add row to dataframe </br>
<br>to add row a Series (containing only one row) is appended to an existing Dataframe </br>

```python
#sort list and check if entry is really gone
df_locations2.sort_values('Postcode_list')
df_locations2
#now add row to dataframe;
#to add row a Series (containing only one row) is appended to an existing Dataframe
df_locations3 = df_locations2.append(pd.Series(['M7R',43.5953,-79.6405], index=df_locations2.columns), ignore_index = True)
df_locations3
```

```python
df_locations4=df_locations3.sort_values('Postcode_list')
df_locations4
```

reset index

```python
#reset index
df_locations4.reset_index()
```

<br>now join both dataframes based on postcode</br>
<br>join via merge, how = inner says its an inner join, left_on/ right_on says via which column</br>
<br> drop column Postcode_list   -    axis = 1 says its the column </br>

```python
#now join both dataframes based on postcode
# join via merge, how = inner says its an inner join, left_on/ right_on says via which column
df_final = pd.merge(df6, df_locations4, how='inner', left_on = 'Postcode', right_on ='Postcode_list')
#drop column Postcode_list
#axis = 1 says its the column
df_final.drop(['Postcode_list'], axis = 1)
```

<br>now go for the clusteranalysis </br>
<br>first display Toronto </br>

```python
import numpy as np
import json # library to handle JSON files
import matplotlib.cm as cm
import matplotlib.colors as colors
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
import folium # map rendering library
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
from sklearn.cluster import KMeans
```

```python
#get the coordinates of Toronto;

address = 'Toronto, ON'
geolocator = Nominatim(user_agent ="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
#map of Toronto
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_final['Latitude'], df_final['Longitude'], df_final['Borough'], df_final['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

<br> after some trying I went for 6 clusters </br>
<br> this displays best the different areas based on geography</br>
<br>base of cluster analysis is Lattitude & Longitude</br>

```python
# set number of clusters
kclusters = 6
#drop all non relevant rows
#clustering is based on euklidian distance therefore you need at least two columns with numbers
toronto_grouped_clustering = df_final.drop(['Postcode','Borough','Neighbourhood', 'Postcode_list'],1)
toronto_grouped_clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)
#list of all cluster labels
kmeans.labels_
```

```python
#del df_final['Cluster Labels']
#df_final
```

<br> adding the cluster labels so that you can see which entry has which Cluster</br>

```python
# add clustering labels
df_final.insert(0, 'Cluster_Labels', kmeans.labels_)
```

```python
df_final
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_final['Latitude'], df_final['Longitude'], df_final['Neighbourhood'], df_final['Cluster_Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

<br> create a small summary for the different clusters</br>

```python
#overview of all clusters
df_overview = df_final.groupby('Cluster_Labels').count()
df_overview=df_overview.drop(['Borough','Neighbourhood','Postcode_list','Latitude','Longitude'],1)
df_overview=df_overview.rename(columns={'Postcode':'count'})
df_overview
```

<br>calculation of the percentage of each Clusters</br>

```python
#calculate percentage;
df_overview['percentage']=df_overview['count']/df_overview['count'].sum() *100
#round result to one decimal
df_overview['percentage']=df_overview['percentage'].apply(lambda x: round(x,1))
df_overview
```
